```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
num_days=365
mean_sales=50
std_dev=10
```

```python
dates=pd.date_range(start='2023-01-01',periods=num_days,freq='D')
```

```python
sales=np.random.normal(loc=mean_sales, scale=std_dev, size=num_days)
```

```python
data = pd.DataFrame( { 'Date' : dates, 'Ventes' : sales   } ) 
```

```python
data.head(20)
```

```python
data['Moyenne_Cumulative'] = data['Ventes'].expanding().mean()
```

```python
data.head(10)
```

```python
data['Moyenne_lissee'] = data['Ventes'].rolling(window=7).mean()
```

```python
plt.subplot(2,1,1)
plt.plot(data['Date'], data['Moyenne_Cumulative'] )
plt.subplot(2,1,2)
plt.plot(data['Date'], data['Ventes'] )
plt.plot(data['Date'], data['Moyenne_lissee'] )
plt.show();
```
