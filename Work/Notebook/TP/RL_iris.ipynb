```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_iris

# Création du modèle de régression linéaire
model = LinearRegression()

# Les données
iris = load_iris()
# print(iris) c'est un dictionnaire 
# print( iris.feature_names) 
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

print( iris_df.head(2) )

X = iris_df[['sepal length (cm)']].values # variable qui explique - indépendante
Y = iris_df['petal length (cm)'].values  # variable cible ou expliquée - dépendante

#print(X)
#print(y)
```

```python
# Entraînement du modèle
model.fit(X, Y) # X et Y sont à déterminées en fonction de vos données

# Coefficient (pente) Y = C*X + I 
coefficient = model.coef_[0]  # valeur C de la pente de notre droite 

# Interception - ordonnée à l'origine 
interception = model.intercept_

print(f"Coefficient : {coefficient}")
print(f"Interception : {interception}")

# Prédictions du modèle == la droite de régression
y_pred = model.predict(X)

# Evaluation du modèle
r_squared = model.score(X, Y)
print(f"Coefficient de détermination R² : {r_squared}")
```
