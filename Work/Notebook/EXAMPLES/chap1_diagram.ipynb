```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)  # Pour des rÃ©sultats reproductibles
num_days = 365
mean_sales = 50  # Ventes moyennes quotidiennes
std_dev = 10  # Ãcart type pour simuler les fluctuations quotidiennes
```

```python
dates = pd.date_range(start='2023-01-01', periods=num_days, freq='D')
sales = np.random.normal(loc=mean_sales, scale=std_dev, size=num_days)
```

```python
dates
```

```python
data = pd.DataFrame({'Date': dates, 'Ventes': sales})
```

```python
data
```

```python
data['Moyenne_Cumulative'] = data['Ventes'].expanding().mean()
```

```python
data['Moyenne_LissÃ©e'] = data['Ventes'].rolling(window=7).mean()
```

```python
plt.figure(figsize=(14, 7))

# Graphique de la moyenne cumulative
plt.subplot(2, 1, 1)
plt.plot(data['Date'], data['Moyenne_Cumulative'], label='Moyenne Cumulative')
plt.xlabel('Date')
plt.ylabel('Ventes Moyenne Cumulative')
plt.title('Moyenne Cumulative des Ventes Quotidiennes')
plt.legend()

# Graphique des ventes quotidiennes et de la moyenne lissÃ©e
plt.subplot(2, 1, 2)
plt.plot(data['Date'], data['Ventes'], label='Ventes Quotidiennes')
plt.plot(data['Date'], data['Moyenne_LissÃ©e'], label='Moyenne LissÃ©e (7 jours)', color='orange')
plt.xlabel('Date')
plt.ylabel('Ventes')
plt.title('Ventes Quotidiennes et Moyenne LissÃ©e')
plt.legend()

plt.tight_layout()
plt.show()
```
