```python
import seaborn as sns
import matplotlib.pyplot as plt

# Charger les donn√©es "tips"
tips = sns.load_dataset("tips")
tips.head()
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.title("Relation entre le montant total de la note et le pourboire en fonction du sexe")
plt.xlabel("Montant total de la note")
plt.ylabel("Pourboire")
plt.show();
```

```python
tips['time'].unique()
```

```python
tipsDinner = tips[ tips[ 'time' ] == 'Dinner' ]
tipsLunch = tips[ tips[ 'time' ] == 'Lunch' ]
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x="total_bill", y="tip", hue="sex", data=tipsDinner)
plt.title("Relation entre le montant total de la note et le pourboire pour le d√Æner en fonction du sexe")
plt.xlabel("Montant total de la note")
plt.ylabel("Pourboire")
plt.show();
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x="total_bill", y="tip", hue="sex", data=tipsLunch)
plt.title("Relation entre le montant total de la note et le pourboire pour le lunch en fonction du sexe")
plt.xlabel("Montant total de la note")
plt.ylabel("Pourboire")
plt.show();
```
