```python
import seaborn as sns
import matplotlib.pyplot as plt

# Chargement du jeu de donnes Titanic depuis Seaborn
titanic = sns.load_dataset("titanic")

# Cration du diagramme en violon
sns.violinplot(x='class', y='age', data=titanic)

# Ajout des labels et du titre
plt.xlabel('Classe de voyage')
plt.ylabel('ge')
plt.title('Distribution des ges par classe de voyage')

# Affichage du graphique
plt.show()
```

```python
children = titanic[titanic['who'] == 'child']
plt.figure(figsize=(10, 6))
sns.violinplot(x='who', y='age', hue='survived', data=children, split=True)
plt.title('Distribution des ges des enfants  bord du Titanic par survie')
plt.xlabel('Catgorie')
plt.ylabel('ge')
plt.legend(title='Survived', loc='upper left')
plt.show()
```
