```python
import seaborn as sns
import matplotlib.pyplot as plt

# Chargement du jeu de données Titanic depuis Seaborn
titanic = sns.load_dataset("titanic")

# Création du diagramme en violon
sns.violinplot(x='class', y='age', data=titanic)

# Ajout des labels et du titre
plt.xlabel('Classe de voyage')
plt.ylabel('Âge')
plt.title('Distribution des âges par classe de voyage')

# Affichage du graphique
plt.show()
```

```python
children = titanic[titanic['who'] == 'child']
plt.figure(figsize=(10, 6))
sns.violinplot(x='who', y='age', hue='survived', data=children, split=True)
plt.title('Distribution des âges des enfants à bord du Titanic par survie')
plt.xlabel('Catégorie')
plt.ylabel('Âge')
plt.legend(title='Survived', loc='upper left')
plt.show()
```
